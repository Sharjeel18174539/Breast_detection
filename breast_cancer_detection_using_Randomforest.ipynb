Breast cancer Detection Using Random Forest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df.diagnosis.value_counts()
```

```python
df.diagnosis.value_counts()/len(df)*100
```

```python
x=df.drop(['id', 'diagnosis'], axis=1)
```

```python
x.head(5)
```

Train Test Split

```python
y=df.diagnosis.values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25)
```

```python
y_train
```

```python
x_train
```

```python
x_test
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train_sc
```

```python
from sklearn.ensemble import RandomForestClassifier
model_randomforest = RandomForestClassifier(n_estimators=100, random_state=42)
model_randomforest.fit(x_train_sc, y_train)
```

```python
y_pred_rd=model_randomforest.predict(x_test_sc)
```

```python
y_pred_rd
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, y_pred_rd) 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_rd)
```

```python
#classification metrics for model behaves

from sklearn.metrics import accuracy_score
print(round(accuracy_score(y_test,y_pred_rd)*100, 2))
```
