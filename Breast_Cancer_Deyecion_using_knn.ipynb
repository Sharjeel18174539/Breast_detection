Breast Cancer Detection Using KNN Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.diagnosis.value_counts()
```

```python
df.diagnosis.value_counts()/len(df)*100
```

now splitting the data in x and y

```python
df.dropna(axis=1,inplace=True)
```

```python
df.head()
```

```python
x=df.drop(['id', 'diagnosis'], axis=1)
```

```python
x
```

```python
y=df.diagnosis.values
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.25)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train_sc
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Initiate the classifier
model_knn=KNeighborsClassifier()

#passing the data to classifier
model_knn.fit(x_train_sc, y_train)
```

```python
y_pred_knn=model_knn.predict(x_test_sc)
```

```python
y_pred_knn
```

```python
y_test
```

```python
#classification metrics for model behaves

from sklearn.metrics import accuracy_score

print(round(accuracy_score(y_test,y_pred_knn)*100, 2))
```
